```python
#Explain the significance of Python keywords and provide examples of five keywords.


"""
Python keywords are reserved words that have predefined meanings in the Python language. 
These keywords cannot be used as identifiers, such as variable names or function names, in Python programs.
Using Python keywords allows the programmer to write code that adheres to the syntax and rules of the language. Here are five examples of Python keywords:

"""

#while:Used for loop execution based on a condition. 
x = 0
while x < 5:
    print(x)
    x += 1
    
    
#def: The `def` keyword is used to define a function in Python.
def greet():
    print("Hello, nice to meet you!")

greet()
    
    
#class: Used to define classes.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

# Example creating an instance of the class
person1 = Person("Amit", 21)
print(person1.greet())


#for: Used for iteration over sequences.
for i in range(5):
    print(i)
    
    
#if: Used for conditional execution
x = 10

if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")

```
